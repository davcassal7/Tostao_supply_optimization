## Inventario Actual

```python
inventario = pd.read_csv(path + 'inventario_actual.csv')
# Se imprimen las primeras 10 filas del DataFrame para verificar que se haya cargado correctamente
inventario.head(10)
```

```python
#Se valida la estructura del DataFrame y los vacíos de las columnas
inventario.info()
```

```python
# Se obtiene el número de tiendas
n_tiendas = inventario.id_tienda.nunique()
print(f"El catálogo contiene {n_tiendas} tiendas únicas.")
```

```python
#Se van a validar cual es el prodcuto con mayor y menor cantidad de inventario por tienda,
#para esto se va a hacer un merge entre el inventario y el catálogo para obtener el nombre del producto

inventario_nombres = inventario.merge(catalogo[['id_producto', 'nombre']], on='id_producto', how='left')
tiendas_stock_stats = (
    inventario_nombres.groupby('id_tienda')
    .agg(
        producto_mayor_stock=('nombre', lambda x: x[inventario_nombres.loc[x.index, 'stock_actual'].idxmax()]),
        mayor_stock=('stock_actual', 'max'),
        producto_menor_stock=('nombre', lambda x: x[inventario_nombres.loc[x.index, 'stock_actual'].idxmin()]),
        menor_stock=('stock_actual', 'min')
    )
    .reset_index()
)

tiendas_stock_stats
```

Hay varias tiendas que en el momento no tienen stock de alguno de los productos:
- STORE_07: Buñuelo
- STORE_08: Capuccino
- STORE_11: Café con Leche
- STORE_17: Tinto
- STORE_19: Pastel de Pollo

```python
#Se revisa el inventario total por producto
inventario_productos = (
    inventario_nombres.groupby('nombre')
    .agg(
        stock_total=('stock_actual', 'sum')
    )
)

print(f'El inventario promedio por producto es: {round(inventario_productos["stock_total"].mean(), 1)}')

inventario_productos.sort_values(by='stock_total', ascending=False).head(10)

```

*El producto con mayor stock entre las tiendas es el Pan de Bono y el de menos es el Café con Leche*

```python
#Se revisan los inventarios por ciudad, para esto se hace un merge entre el inventario y las tiendas para obtener la ciudad de cada tienda
inventario_ciudad = inventario_nombres.merge(tiendas[['id_tienda', 'ciudad']], on='id_tienda', how='left')

inventario_ciudad_stats = (
    inventario_ciudad.groupby(['ciudad', 'nombre'])
    .agg(
        stock_total=('stock_actual', 'sum'),
        stock_promedio=('stock_actual', 'mean'),
        stock_maximo=('stock_actual', 'max'),
        stock_minimo=('stock_actual', 'min')
    )
)

inventario_ciudad_stats.sort_values(by='stock_total', ascending=False)
```

- En Bogota el producto con mayor stock es el Cappuccino, y el pastel de pollo es el de menor stock
- En Medellín y Cali el Jugo de Naranaja presenta el mayor stock mientras que el Capucciono tiene el menor stock para Medellín y el Buñuelo en Cali


## Ventas Históricas

```python
ventas = pd.read_csv(path + 'ventas_historicas.csv')
# Se imprimen las primeras 10 filas del DataFrame para verificar que se haya cargado correctamente
ventas.head(10)
```

```python
#Se valida la estructura del DataFrame y los vacíos de las columnas
ventas.info()
```

```python
# Se convierte el campo fecha a tipo datetime
ventas['fecha'] = pd.to_datetime(ventas['fecha'])

min_fecha = ventas['fecha'].min()
max_fecha = ventas['fecha'].max()
num_dias = ventas['fecha'].nunique()
num_semanas = ventas['fecha'].dt.to_period('W').nunique()


print(f"""se tienen ventas para el periodo comprendido entre {min_fecha.date()} y {max_fecha.date()}.
Esto equivale a un total de {num_dias} días y {num_semanas} semanas
""")  
```

```python
# Se van a revisar las ventas por ciudad, para esto se hace un merge entre las ventas y las tiendas para obtener la ciudad de cada tienda
# Además se hace un merge con el catálogo para obtener el nombre del producto vendido

ventas_ciudad = ventas.merge(tiendas[['id_tienda', 'ciudad']], on='id_tienda', how='left')\
    .merge(catalogo[['id_producto', 'nombre']], on='id_producto', how='left')

ventas_ciudad_stats = (
    ventas_ciudad.groupby(['ciudad', 'nombre'])
    .agg(
        ventas_totales=('unidades_vendidas', 'sum'),
        ventas_promedio=('unidades_vendidas', 'mean'),
        ventas_maximas=('unidades_vendidas', 'max'),
        ventas_minimas=('unidades_vendidas', 'min')
    )
)

ventas_ciudad_stats.sort_values(by='ventas_totales', ascending=False)
```

```python
#Se cuentan cuantas ventas (días de venta) se tienen por producto y ciudad
ventas_diarias = (
    ventas_ciudad.groupby(['ciudad', 'nombre'])
    .agg(
        dias_venta=('fecha', 'nunique')
    )
)

ventas_diarias
```

# Análisis ventas por semana

```python
# Se unen las fuentes de datos para obtener un DataFrame con la información completa
data = ventas.merge(catalogo, on='id_producto', how='left').merge(tiendas, on='id_tienda', how='left').merge(inventario, on=['id_tienda', 'id_producto'], how='left')

# Se convierte la columna 'fecha' a tipo datetime para facilitar el análisis temporal
data['fecha'] = pd.to_datetime(data['fecha'])

# Se crea la columna de semana año para agrupar las ventas por semana
data['semana'] = data['fecha'].dt.isocalendar().week

# Se imprimen las primeras 10 filas del DataFrame para verificar que se haya cargado correctamente
data.head(10)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Se revisa la evolución en el tiempo de las ventas por producto y ciudad
ventas_semana = (
    data.groupby(['semana', 'ciudad', 'nombre'])
    .agg(
        ventas_semanales=('unidades_vendidas', 'sum')
    )
    .reset_index()
)


ciudades = sorted(ventas_semana['ciudad'].unique())
productos = sorted(ventas_semana['nombre'].unique())

# Se realiza un gráfico que muestra las ventas semanales en el tiempo y las compara con posibles outlier, usando el rango intercuartil
fig = make_subplots(
    rows=len(ciudades),
    cols=1,
    subplot_titles=ciudades,
    shared_xaxes=True
)

# Diccionario para rastrear la relación de cada traza con su producto
trace_product_map = []
primer_producto = productos[0]

# Se agregan los trazos de las ventas y los límiter superiro e inferior
for producto in productos:
    es_primer_prod = (producto == primer_producto)

    for c_idx, ciudad in enumerate(ciudades, start=1):
        df = ventas_semana[
            (ventas_semana['ciudad'] == ciudad) &
            (ventas_semana['nombre'] == producto)
        ].sort_values('semana')
        
        # Cálculo de Outliers (IQR)
        q1 = df['ventas_semanales'].quantile(0.25)
        q3 = df['ventas_semanales'].quantile(0.75)
        ric = q3 - q1
        limite_inferior = max(0, q1 - 1.5 * ric)
        limite_superior = q3 + 1.5 * ric

        # Ventas Semanales
        fig.add_trace(
            go.Scatter(
                x=df['semana'],
                y=df['ventas_semanales'],
                mode='lines+markers',
                name=producto,
                legendgroup=producto,
                showlegend=(c_idx == 1),
                visible=es_primer_prod
            ),
            row=c_idx, col=1
        )
        trace_product_map.append(producto)

        # Límite Superior
        fig.add_trace(
            go.Scatter(
                x=df['semana'],
                y=[limite_superior] * len(df),
                mode='lines',
                line=dict(dash='dash', width=1.5, color='red'),
                name=f'Lím. Sup. ({producto})',
                legendgroup=producto,
                showlegend=False,
                hoverinfo='text',
                hovertext=f'Lím. Superior {producto}: {limite_superior:.1f}',
                visible=es_primer_prod
            ),
            row=c_idx, col=1
        )
        trace_product_map.append(producto)

        # Límite Inferior
        fig.add_trace(
            go.Scatter(
                x=df['semana'],
                y=[limite_inferior] * len(df),
                mode='lines',
                line=dict(dash='dash', width=1.5, color='orange'),
                name=f'Lím. Inf. ({producto})',
                legendgroup=producto,
                showlegend=False,
                hoverinfo='text',
                hovertext=f'Lím. Inferior {producto}: {limite_inferior:.1f}',
                visible=es_primer_prod
            ),
            row=c_idx, col=1
        )
        trace_product_map.append(producto)

# Se configuran los botones del menú desplegable
buttons = []
for producto in productos:
    visible = [p == producto for p in trace_product_map]
    buttons.append(
        dict(
            label=producto,
            method='update',
            args=[{'visible': visible}, {'title': f'Ventas semanales y outliers - {producto}'}]
        )
    )

# Se ajusta el layout final
fig.update_layout(
    title=f'Ventas Semanales y Outliers - {primer_producto}',
    updatemenus=[dict(buttons=buttons, direction='down', x=0.0, y=1.05, showactive=True)],
    height=320 * len(ciudades),
    width=950,
    legend_title='Producto'
)

fig.update_xaxes(title_text='Semana')
fig.update_yaxes(title_text='Ventas semanales')

fig
```

Análisando las gráficas para cada uno de los 8 productos se encuentra que:
1. Buñuelos: 
- En la **semana 2**, para **Bogotá** se tomaron valores por debajo del límite inferior (posible outlier de un valor muy bajo para la ciudad)
- En la **semana 9**, para **Cali** se tomaron valores por encima del límite superior (posible outlier de un valor muy alto para la ciudad)
2. Café con Leche:
- En la **semana 11**, para **Bogotá** se tomaron valores por encima del límite superior (posible outlier de un valor muy alto para la ciudad)
3. Capuccino:
- En las **semanas 5 y 12**, para **Bogotá** se tomaron valores por debajo del límite inferior (posibles outliers de un valor muy bajo para la ciudad)
4. Croissant: No se evindencian posibles outliers  

5. Jugo de Naranja: No se evindencian posibles outliers  
6. Pan de Bono: No se evindencian posibles outliers  
7. Pasel de Pollo:
- En la **semana 13**, para **Bogotá** se tomaron valores cercanos del límite superior (posible outlier de un valor muy alto para la ciudad)
8. Tinto: No se evindencian posibles outliers. Sin embargo, en **Medellín** en la semana 3 se observan valores cercanos al límite superior, siendo un comportamiento fuera del normal.
